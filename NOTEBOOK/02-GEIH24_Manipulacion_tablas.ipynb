```python
#EL MERGE NOS AYUDA PARA HACER UN JOIN PARA JUNTAR UNA COLUMNA A UNA TABLA 
df_ocupados_genero=pd.merge(df_ocupados, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271']], on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
#calcular el total de mujeres y hombres ocupadas 
OH= df_ocupados_genero[df_ocupados_genero['P3271']==1]['FEX_C18'].sum()
OM= df_ocupados_genero[df_ocupados_genero['P3271']==2]['FEX_C18'].sum()
print(f'Hombres en empleo = {int(OH)}')
print(f'Mujeres en empleo = {int(OM)}')
```

```python
df_desocupados_genero=pd.merge(df_desocupados, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271']], on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
#calcular el total de mujeres y hombres desocupadas 
DH= df_desocupados_genero[df_desocupados_genero['P3271']==1]['FEX_C18'].sum()
DM= df_desocupados_genero[df_desocupados_genero['P3271']==2]['FEX_C18'].sum()
print(f'Hombres en desempleo = {int(DH)}')
print(f'Mujeres en desempleo = {int(DM)}')
```

```python
df_fuerza_genero=pd.merge(df_fuerza, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271']], on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
#calcular el total de mujeres y hombres en fuerza de trabajo  
DH= df_fuerza_genero[df_fuerza_genero['P3271']==1]['FEX_C18'].sum()
DM= df_fuerza_genero[df_fuerza_genero['P3271']==2]['FEX_C18'].sum()
print(f'Hombres en edad de trabajar = {int(DH)}')
print(f'Mujeres en edad de trabajar = {int(DM)}')
```

```python
PDQH= df_desocupados_genero[(df_desocupados_genero['DPTO']==63)& (df_desocupados_genero['P3271']==1)]['FEX_C18'].sum()
PDQM= df_desocupados_genero[(df_desocupados_genero['DPTO']==63)& (df_desocupados_genero['P3271']==2)]['FEX_C18'].sum()
print(f'Hombres desmepleados del quindío: {PDQH}')
print(f'Mujeres desempleados del quindío: {PDQM}')
```
